### *1 - Description d'un notebook Jupyter*

Dans un notebook Jupyter, il existe 3 sortes de *cellules*, c'est à dire des cases pour entrer des éléments.
- des cellules de type ***Markdown*** pour avoir simplement du texte, formatable avec quelques signes
- des cellules de type ***Raw*** pour montrer une commande, mais qui n'est pas exécutable  
- des cellules de type ***Code***, pour contenir des commandes exécutables

### *2 - Utilisation d'un notebook Jupyter*
    
Un notebook Jupyter est flexible, vous pouvez ajouter, modifier, déplacer les cellules à volonté.  

Pour **modifier** une cellule, double-cliquez dessus.    

Pour **ajouter** une nouvelle cellule, cliquez sur l'icône <kbd>➕</kbd> dans la barre des menus.  

Pour **bouger** une cellule, vous pouvez la "cliquer-glisser" vers le haut ou vers le bas.  

Pour **choisir ou modifier le type de cellule**, vous cliquez dessus (la barre bleue s'affiche à gauche de la cellule sélectionnée), et vous choisissez le type voulu dans le menu déroulant dans la barre des menus.  

Pour **exécuter une cellule**, c'est à dire voir son effet :
- la combinaison de touches <kbd>Ctrl</kbd> + <kbd>Enter</kbd> exécute une cellule.  
- la combinaison de touches <kbd>Shift</kbd> + <kbd>Enter</kbd> exécute une cellule puis passe à la suivante.   
- la combinaison de touches <kbd>Alt</kbd> + <kbd>Enter</kbd> exécute une cellule puis en crée une nouvelle (vide) en dessous. 

Vous pouvez aussi utiliser l'icône <kbd>▶︎</kbd> en haut du notebook.  

    
L'**exécution d'une cellule** va permettre de voir son effet :  
- une cellule *Markdown* prendra son aspect formaté   
- une cellule *Raw* prendra son aspect de cadre gris   
- une cellule *Code* lancera la ou les commandes indiquée(s).
